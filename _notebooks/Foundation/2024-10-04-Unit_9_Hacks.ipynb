```java
// 9.1 and 9.3 Hacks

public class Shape {
    protected String name;
    protected int length;
    protected int width;

    public Shape() {
        this.name = "Shape";
        this.length = 10;
        this.width = 5;
    }

    public Shape(String name, int length, int width) {
        this.name = name;
        this.length = length;
        this.width = width;
    }

    public double area() {
        return this.length * this.width;
    }

    public double perimeter() {
        return 2 * (length + width);
    }

    public void print_something() {
        System.out.println("This is a shape");
    }
}

public class Circle extends Shape {
    private double radius;

    public Circle() {
        super("Circle", 0, 0);
        this.radius = 5.0;
    }

    public Circle(String name, double radius) {
        super(name, 0, 0);
        this.radius = radius;
    }

    @Override
    public double area() {
        return Math.PI * Math.pow(radius, 2);
    }

    @Override
    public double perimeter() {
        return 2 * Math.PI * radius;
    }

    @Override
    public void print_something() {
        super.print_something();
        System.out.println("This is a shape and also a circle with radius " + radius);
    }

    public String toString() {
        return name + " has an area of " + area() + " units and a perimeter of " + perimeter() + " units";
    }
}

public class Hexagon extends Shape {
    private int side;

    public Hexagon() {
        super("Hexagon", 0, 0);
        this.side = 5;
    }

    public Hexagon(String name, int side) {
        super(name, 0, 0);
        this.side = side;
    }

    @Override
    public double area() {
        return (3 * Math.sqrt(3) * Math.pow(side, 2)) / 2;
    }

    @Override
    public double perimeter() {
        return 6 * side;
    }

    @Override
    public void print_something() {
        super.print_something();
        System.out.println("This is a shape and also a hexagon with side length " + side);
    }

    public String toString() {
        return name + " has an area of " + area() + " units and a perimeter of " + perimeter() + " units";
    }
}

public class Ellipse extends Shape {

    public Ellipse() {
        super("Ellipse", 10, 5);
    }

    public Ellipse(String name, int length, int width) {
        super(name, length, width);
    }

    @Override
    public double area() {
        return Math.PI * (length / 2.0) * (width / 2.0);
    }

    @Override
    public double perimeter() {
        double a = length / 2.0;
        double b = width / 2.0;
        return Math.PI * (3 * (a + b) - Math.sqrt((3 * a + b) * (a + 3 * b)));
    }

    @Override
    public void print_something() {
        super.print_something();
        System.out.println("This is a shape and also an ellipse with semi-axes " + (length / 2.0) + " and " + (width / 2.0));
    }

    public String toString() {
        return name + " has an area of " + area() + " units and a perimeter of " + perimeter() + " units";
    }
}

public class Main {
    public static void main(String[] args) {
        Shape shape = new Shape();
        shape.print_something();
        System.out.println("Area of default shape: " + shape.area());
        System.out.println("Perimeter of default shape: " + shape.perimeter());

        Circle circle = new Circle("Circle", 7.5);
        circle.print_something();
        System.out.println(circle);

        Hexagon hexagon = new Hexagon("Hexagon", 10);
        hexagon.print_something();
        System.out.println(hexagon);

        Ellipse ellipse = new Ellipse("Ellipse", 12, 8);
        ellipse.print_something();
        System.out.println(ellipse);
    }
}

Main.main(null);
```

```java
class Book {
    public String title;
    public double price;

    public Book(String title, double price) {
        this.title = title;
        this.price = price;
    }

    public String getBookInfo() {
        return "Title: " + title + ", Price: $" + price;
    }
}

class Textbook extends Book {
    private int edition;

    public Textbook(String title, double price, int edition) {
        super(title, price);
        this.edition = edition;
    }

    public int getEdition() {
        return this.edition;
    }

    public boolean canSubstituteFor(Textbook other) {
        return this.title.equals(other.title) && this.edition >= other.edition;
    }

    @Override
    public String getBookInfo() {
        return super.getBookInfo() + ", Edition: " + edition;
    }

    @Override
    public String toString() {
        return "The title of the book is " + this.title + " with a price of $" + this.price + " and with an edition of " + this.edition;
    }
}

public class Main {
    public static void main(String[] args) {
        Textbook sheeshing = new Textbook("Life of Pi", 75.0, 10);
        Textbook sheesh = new Textbook("Life of Pi", 74.0, 11);

        System.out.println(sheeshing.getBookInfo());
        System.out.println(sheesh.getBookInfo());

        System.out.println(sheeshing.canSubstituteFor(sheesh));
        System.out.println(sheesh.canSubstituteFor(sheeshing));

        System.out.println(sheeshing);
        System.out.println(sheesh);
    }
}

Main.main(null);
```
